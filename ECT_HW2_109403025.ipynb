```python
# 1.
import pandas as pd
data = pd.read_csv('dataset.csv')
data.head()
```

```python
# 2.
data.isnull().sum() # 計算dataset.csv中各欄位內的空值
```

```python
data.info()
```

```python
# 3.
data = data.loc[data['year'] > 2004] # 將2004年前的數據刪除
data.tail()
```

```python
# 4.
data = data.dropna(subset=['eco_freedom']) # 移除eco_freedom的空值
data.tail()
```

```python
# 5.
# 用 map() 將欄位轉為數字型態 HIGH=2 NORMAL=1 LOW=0 
data['freedom_own_foreign_currency'] = data['freedom_own_foreign_currency'].map({'HIGH':2,'NORMAL':1,'LOW':0}, na_action='ignore')
data.info()
```

```python
# 6.
data.loc[data['eco_freedom'] < 6 , 'eco_freedom'] = 0 # eco_freedom的值小於6設定為0
data.loc[data['eco_freedom'] >=6 , 'eco_freedom'] = 1 # eco_freedom的值大於等於6設定為1
data = data.rename({'eco_freedom':'freedom'},axis=1)
data.head()
```

```python
# 7.
data = data.fillna(data.mean(numeric_only=True))
```

```python
# 8.
X = data.drop(['year','iso_code','countries','rank','freedom'],axis=1)
X.info()
# X 為訓練資料集
```

```python
y = data['freedom']
# y為測試資料集
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.5, random_state = 12, stratify=y)
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(max_leaf_nodes=5,criterion='gini')
model = model.fit(X_train, y_train)
# 訓練model
```

```python
from sklearn import tree
import matplotlib.pyplot as plt
fn = list(X.columns.values)
cn = ['freedom']
plt.subplots(figsize = (10,10))
tree.plot_tree(model, feature_names=fn,filled=True) # 繪製 decision tree
```

```python
# 9.
from sklearn.metrics import accuracy_score
y_predict = model.predict(X_test) # 使用 X_test 預測分類
accuracy = accuracy_score(y_test, y_predict) # 計算準確率
```

```python
# 10.
from sklearn import metrics
predicted = model.predict(X_test) # 使用X_test預測
names = ['non-freedom', 'freedom']
print(metrics.classification_report(y_test,predicted, target_names=names))
```

```python
# 11.
split_proportion = []
score = []
for i in range(5,10,1):
    X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = float(i/10), random_state = 12, stratify=y)
    model = DecisionTreeClassifier(max_leaf_nodes=5,criterion='gini')
    model = model.fit(X_train, y_train)
    # 訓練模型
    y_predict = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_predict)
    # 準確率
    split_proportion += [str(i*10)+'/'+str((10-i)*10)] # 切割的比例
    score += [accuracy] # 各比例的準確率
    
decision_tree = pd.DataFrame({'split_proportion':split_proportion,'score':score})
print(decision_tree)
```

```python
# 12. 繪製折線圖
import matplotlib.pyplot as plt
plt.plot(split_proportion,score)
plt.title('train/test split proportion & result')
plt.legend(['score'])
plt.show()
```
